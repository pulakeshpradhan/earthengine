```r
# Load the necessary libraries
library(readxl)
library(dmetar)
library(meta)
library(tidyverse)
library(robvis)
library(metafor)
library(metasens)

# Set the working directory (if necessary)
# setwd("D:/EndNote")

# Import the data
dat <- read_excel("/content/data.xlsx")
print(dat)

# Pool the prevalence data using a GLMM and logit-transformed proportions
m.prop <- metaprop(event = event,
                   n = sample,
                   studlab = study,
                   data = dat,
                   method = "GLMM",
                   sm = "PLOGIT",
                   fixed = FALSE,
                   random = TRUE,
                   hakn = TRUE,
                   title = "")
print(summary(m.prop))

# Create a forest plot
forest(m.prop,
       sortvar = study,
       prediction = TRUE,
       print.tau2 = FALSE,
       leftlabs = c("Author", "Events", "N"))
```

```r
forest(m.prop,
            sortvar = study,
            prediction = TRUE,
            print.tau2 = FALSE,
            leftlabs = c("Author", "Events", "N"))
```

```r
#Layout RevMan5
png(file = "forest.png", width = 3200, height = 4000, res = 300)
forest(m.prop, prediction = TRUE, common = FALSE,
            layout = "RevMan5",
            text.random = "Pooled prevalence (REM)",
            smlab = "",
            pooled.events = T)
dev.off()
```

```r
#Leave-one-out analysis
png(file = "Leave_forest.png", width = 3200, height = 4000, res = 300)
forest(metainf(m.prop, pooled = "random"),
            prediction = TRUE, common = FALSE,
            layout = "RevMan5",
            text.random = "Pooled prevalence (REM)",
            smlab = "",
            pooled.events = T,
            text.addline1 = "Sensitivity analysis",
            text.addline2 = "Leave-one-out meta-analysis",
            xlim = c(0.4, 0.75))
dev.off()
```

```r
#Forestplot after removing outliars
m1<-update(m.prop, exclude = c(5))
m1

png(file = "forestplot_update.png", width = 2800, height = 2400, res = 300)
forest(m1, prediction = TRUE,layout = "RevMan5")
dev.off()
```

```r
#Regression, Bubble plot

tiff(filename = "bubble.tiff", res = 175, width = 1600, height = 1000)
metta.rma <- rma(yi = m.prop$TE, sei = m.prop$seTE, method = m.prop$method.tau, mods = dat$n)
summary(metta.rma)
regplot(x = metta.rma, pi = T, legend = TRUE, xlab = "Sample size", ylab = "Proportion of cases (transformed)", col = "brown", lcol = c("red","green","blue","black"), label ="piout" )
title("Bubble plot demonstrating meta-regression based upon sample size")
dev.off()
```

```r
#DOI Plot
png(file = "DOI.png", width = 3000, height = 2400, res = 300)
doiplot(m.prop,
        xlim=c(-0.2, 0.5),
        xlab = "ES",
  ylab = "|Z-score|",
  lfkindex = TRUE,
  pos.lfkindex = "topleft")
dev.off()
```

```r
#Funnel Plot
png(file = "Funnel.png", width = 3000, height = 2400, res = 300)
funnel(m.prop,
            xlim = c(-0.5, 1.0),
            studlab = F, linreg = TRUE)
dev.off()
```

```r
#GHOSH Plot
m.rma <- rma(yi = m.prop$TE,
             sei = m.prop$seTE,
             method = m.prop$method.tau,
             test = "knha")

res.gosh <- gosh(m.rma, subset=20000)
```

```r
#Plot GHOSH
png(file = "ghosh.png", width = 3000, height = 2400, res = 300)
plot(res.gosh, alpha = 0.01)
dev.off()


res.gosh.diag <- gosh.diagnostics(res.gosh,
                                  km.params = list(centers = 2),
                                db.params = list(eps = 0.08,
                                                   MinPts = 50))
res.gosh.diag

plot(res.gosh.diag)
```
